```python
def Adj_df(stock):
  
    category = ['F5_TAlibAll', 'F5_Future', 'F5_Category', 'F5_TAlib15', 'F5_Nothing']
    change_name = ['LSTM_F5_TI189','LSTM_F5_Futures','LSTM_F5_Category','LSTM_F5_XGBTI15','LSTM_F5_Nothing']
    All_df = []

    for i in range(len(category)):

      p = f'/content/drive/MyDrive/Final Project/_4_Build Models/_Final_Models/Model Evaluate Data/LSTM & GRU {category[i]}/researchdata/{stock}_Prediction_Data.csv'
      df = pd.read_csv(p)

      # 改欄位名稱 
      a = change_name[i][5:]
      df.rename(columns={'prediction_LSTM': change_name[i] ,'prediction_GRU':f'GRU_{a}'}, inplace=True)

      # 取test的部分
      df = df[df.split == 'test'].drop(['split'],axis=1)
      
      All_df.append(df)

    # 合併表格

    M = All_df[-1]

    for i in range(4):

      df = All_df[i]
      M = pd.merge(M,df.drop(['historical'],axis=1), how='left',on='date')

      #print('=====> ',stock,M.shape,M.isnull().sum())

    return M

#print(M)
```

```python
# 取得所有股票代碼
from glob import glob

path = glob('/content/drive/MyDrive/Final Project/_4_Build Models/_Final_Models/Model Evaluate Data/LSTM & GRU F5_Category/researchdata/*_Prediction_Data.csv')

for i in range(len(path)):

  stock = path[i].replace('_','/').split('/')[-3]

  # 修改表格
  f = Adj_df(stock)

  #存檔
  f.to_csv(f'/content/drive/MyDrive/Final Project/_5_Display on Website/折線圖/data_折線/{stock}.csv',index=False)
```
