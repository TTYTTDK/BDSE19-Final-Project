```python
def Adj_df(stock):
    category = ['LSTM_AllFeature','LSTM_Future', 'LSTM_CategoryClose', 'LSTM_TAlib15', 'LSTM_only5']
    change_name = ['LSTM_F5_TI189','LSTM_F5_Futures','LSTM_F5_Category','LSTM_F5_XGBTI15','LSTM_F5_Nothing']
    All_df = []

    for i in range(len(category)):
      
      df = pd.read_csv(f'/content/drive/MyDrive/Final Project/LSTM/v7 data/{category[i]}/researchdata/{stock}_Prediction_Data.csv')
      
      # 改欄位名稱 
      a = change_name[i][5:]
      df.rename(columns={'prediction_LSTM': change_name[i] ,'prediction_GRU':f'GRU_{a}'}, inplace=True)
      
      # 取test的部分
      df = df[df.split == 'test'].drop(['split'],axis=1)
      
      All_df.append(df)

    # 合併表格

    M = All_df[-1]

    for i in range(4):
      df = All_df[i]
      M = pd.merge(M,df, how='left')

    return M

#print(M)
```

```python
# 取得所有股票代碼
from glob import glob
path = glob('/content/drive/MyDrive/Final Project/LSTM/v7 data/LSTM_Future/researchdata/*_Prediction_Data.csv')

for i in range(len(path)):
  stock = path[i].replace('_','/').split('/')[10]

  # 修改表格
  f = Adj_df(stock)

  #存檔
  #f.to_csv(f'/content/drive/MyDrive/Final Project/折線圖/折線_data/{stock}.csv',index=False)



print(f)
```
