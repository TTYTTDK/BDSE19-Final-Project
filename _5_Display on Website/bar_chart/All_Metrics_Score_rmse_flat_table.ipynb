```python
import pandas as pd
import numpy as np
#import matplotlib
#import matplotlib.pyplot as plt
#import seaborn as sns
```

```python
#讀5個檔, 設為muitiindex, 只取RMSE值
A = pd.read_csv('All_Metrics_Score_Time_AllFeature.csv').set_index(['name','modelname','split']).RMSE.xs('Test', level = 'split').to_frame().rename(columns = {'RMSE':'A_RMSE'})
B = pd.read_csv('All_Metrics_Score_Time_TAlib.csv').set_index(['name','modelname','split']).RMSE.xs('Test', level = 'split').to_frame().rename(columns = {'RMSE':'B_RMSE'})
C = pd.read_csv('All_Metrics_Score_Time_Future.csv').set_index(['name','modelname','split']).RMSE.xs('Test', level = 'split').to_frame().rename(columns = {'RMSE':'C_RMSE'})
D = pd.read_csv('All_Metrics_Score_Time_CategoryClose.csv').set_index(['name','modelname','split']).RMSE.xs('Test', level = 'split').to_frame().rename(columns = {'RMSE':'D_RMSE'})
E = pd.read_csv('All_Metrics_Score_Time_Five.csv').set_index(['name','modelname','split']).RMSE.xs('Test', level = 'split').to_frame().rename(columns = {'RMSE':'E_RMSE'})
#3種模型RMSE值合併為一個dataframe
#ALL_rmse = pd.concat([A ,B, C, D, E], axis = 1)
#ALL_rmse.to_csv()
# 畫圖
#final_chart = ALL_rmse.unstack(level=1).plot(kind='bar', title = 'ALL_rmse', figsize = (20,6), fontsize='xx-large').legend(loc = 2)
#final_chart
```

```python
# 分別把5個dataframe都降下一個index做扁平化, 並重新命名欄位名稱以利後續分辨欄位
A1 = A.unstack('modelname').rename(columns = {'GRU':'GRU_F5_TI189', 'LSTM':'LSTM_F5_TI189'})
A1.columns = A1.columns.droplevel()
A1 = A1.rename_axis(columns=None)

B1 = B.unstack('modelname').rename(columns = {'GRU':'GRU_F5_XGBTI15', 'LSTM':'LSTM_F5_XGBTI15'})
B1.columns = B1.columns.droplevel()
B1 = B1.rename_axis(columns=None)

C1 = C.unstack('modelname').rename(columns = {'GRU':'GRU_F5_Futures', 'LSTM':'LSTM_F5_Futures'})
C1.columns = C1.columns.droplevel()
C1 = C1.rename_axis(columns=None)

D1 = D.unstack('modelname').rename(columns = {'GRU':'GRU_F5_Category', 'LSTM':'LSTM_F5_Category'})
D1.columns = D1.columns.droplevel()
D1 = D1.rename_axis(columns=None)

E1 = E.unstack('modelname').rename(columns = {'GRU':'GRU_F5_Nothing', 'LSTM':'LSTM_F5_Nothing'})
E1.columns = E1.columns.droplevel()
E1 = E1.rename_axis(columns=None)

All1 = pd.concat([A1, B1, C1, D1, E1], axis = 1)
All1 = All1.rename(index={'1301.TW':'台塑(1301)','1303.TW':'南亞(1303)','1304.TW':'台聚(1304)','1305.TW':'華夏(1305)',\
                   '1308.TW':'亞聚(1308)','1309.TW':'台達化(1309)','1312.TW':'國喬(1312)','1313.TW':'聯成(1313)',\
                   '1314.TW':'中石化(1314)','1326.TW':'台化(1326)','2002.TW':'中鋼(2002)','2006.TW':'東和鋼鐵(2006)',\
                   '2010.TW':'春源(2010)','2014.TW':'中鴻(2014)','2023.TW':'燁輝(2023)','2027.TW':'大成鋼(2027)',\
                   '2028.TW':'威致(2028)','2030.TW':'彰源(2030)','2031.TW':'新光鋼(2031)','2038.TW':'海光(2038)',\
                   '2603.TW':'長榮(2603)','2605.TW':'新興(2605)','2606.TW':'裕民(2606)','2609.TW':'陽明(2609)',\
                   '2615.TW':'萬海(2615)','2617.TW':'台航(2617)','2636.TW':'台驊控股(2636)','2637.TW':'慧洋-KY(2637)',\
                   '5608.TW':'四維航(5608)','5609.TWO':'中菲行(5609)'})
All1.to_csv('All_Metrics_Score_rmse_flat_table.csv')
```
