```python
# 匯入套件
import pandas as pd
```

```python
# 匯入已知資料
df_ER = pd.read_csv('ER201001202105.csv')
df_ER = df_ER.rename(columns = {'期間':'Date', '新台幣NTD/USD':'NTD/USD'}, inplace = False)
df_ER
```

```python
# 匯入套件
import requests
from bs4 import BeautifulSoup

# 美麗的湯類別
html = requests.get('https://www.cbc.gov.tw/tw/lp-645-1-1-60.html')
html.encoding = 'utf8'
soup = BeautifulSoup(html.text, 'lxml')

# 抓取所需要的資料
items = soup.find('table', class_='rwd-table').find_all('span')
```

```python
# 將資料存入 list
outlist = []
inlist = []

for i in range(4, 46):
    if i%2 == 0:
        inlist.append(items[i].text[0:4]+items[i].text[5:7]+items[i].text[8:10])
    if i%2 == 1:
        inlist.append(items[i].get_text())
        outlist.insert(0, inlist)
        inlist = []
```

```python
# 將資料存入 dataframe
df = pd.DataFrame(outlist, columns = ['Date', 'NTD/USD'])
df_ER = df_ER.append(df, ignore_index = True)
df_ER
```

```python
# 存檔
import os
if os.path.exists('data') is False:
    os.makedirs('data')
    
df_ER.to_csv('data/ExchangeRate_NTD2USD.csv')
```
