```python
# 匯入套件
import pandas as pd
import datetime
import time
import os
```

```python
# 自訂 yahoo finance 爬蟲函式 
def crawl_yahoo_finance_usa(ticker, Y1, M1, D1, Y2, M2, D2):
    period1 = int(time.mktime(datetime.datetime(Y1,M1,D1,8,0).timetuple()))
    period2 = int(time.mktime(datetime.datetime(Y2,M2,D2,8,0).timetuple()))
    query_string = f'https://query1.finance.yahoo.com/v7/finance/download/{ticker}?period1={period1}&period2={period2}&interval=1d&events=history&includeAdjustedClose=true'
    return pd.read_csv(query_string)
```

```python
# 要抓的股票 {'代碼':'名稱'}
stocks = {'%5ETWII':'WeightIndex'}
```

```python
# 建立檔案存檔位置
if os.path.exists('data') is False:
    os.makedirs('data')
```

```python
for tickno, tickname in stocks.items():
    
    # 爬取資料
    df = crawl_yahoo_finance_usa(tickno, 2010, 1, 1, 2021, 6, 30)
    
    # 存檔
    df.to_csv(f'data/Stock_TW_{tickname}.csv', index_label=False)
    print(f'{tickname} was crawled.')
    time.sleep(2)
```
