```python
# 匯入套件
import pandas as pd
import os
import re
```

```python
# list 出 rawdata 下面 Futures 開頭的檔案
filenames = [f for f in os.listdir('rawdata') if re.search('^Futures', f)]
filenames
```

```python
for name in filenames:
    
    # 讀入檔案並將 row 換成從遠至近
    df = pd.read_csv(f'rawdata/{name}')
    df = df.iloc[::-1].reset_index(drop=True)
    
    print(name)
    print(df)
    
    # 存檔
    df.to_csv(f'data/{name}', index_label=False )
    print(f'{name} was saved in data folder')
    print('-'*50)
```
