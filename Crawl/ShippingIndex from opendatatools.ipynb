```python
# 匯入套件
from opendatatools import shippingindex
import pandas as pd
import time
import os
```

```python
# 建立檔案存檔位置
if os.path.exists('data') is False:
    os.makedirs('data')
```

```python
# 查詢 shippingindex 裡面有的內容
df_index, msg = shippingindex.get_index_list()
df_index
```

```python
for name in list(df_index['index']): # ['BDI', 'BPI', 'BCI', 'BSI', 'BDTI', 'BCTI']
    
    # 歷史資料選取
    df, msg = shippingindex.get_index_data(name)
    df = df[df['DATADATE'].between('2010-01-01', '2021-06-30')].reset_index(drop=True)
    
    print(name)
    print(df)
    
    # 存檔
    df.to_csv(f'data/ShippingIndex_{name}.csv', index_label=False)
    print(f'{name} was saved.')
    print('-'*50)   
```
