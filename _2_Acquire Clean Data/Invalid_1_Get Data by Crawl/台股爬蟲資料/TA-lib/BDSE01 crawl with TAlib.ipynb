```python
import numpy as np
import talib
from talib import abstract
import pandas as pd
df_1 = pd.read_csv("DateAdj_Futures_Gold.csv")
```

```python
df_1 = df_1.set_index('Date')
```

```python
df_1
```

```python
df_1.rename(columns={'Price': 'Close', 'Vol.':'Volume'}, inplace=True)
df_1
```

```python
df_1.columns = df_1.columns.map(lambda x:x.lower())
```

```python
df_1 = df_1.astype('float')

# 這裡示範全部 158 種技術指標
ta_list = talib.get_functions()

# 迴圈執行，看看結果吧！
for x in ta_list:
    try:
        # x 為技術指標的代碼，透過迴圈填入，再透過 eval 計算出 output
        output = eval('abstract.'+x+'(df_1)')
        # 如果輸出是一維資料，幫這個指標取名為 x 本身；多維資料則不需命名
        output.name = x.lower() if type(output) == pd.core.series.Series else None
        # 透過 merge 把輸出結果併入 df DataFrame
        df_1 = pd.merge(df_1, pd.DataFrame(output), left_on = df_1.index, right_on = output.index)
        df_1 = df_1.set_index('key_0')
    except:
        print(x)
```

```python
df_1.to_csv('DateAdj_Futures_Gold_TAlib.csv')
#df_2 = pd.read_csv("Stock_TW_2023.TW.TAlib.csv", index_col = 0)
#df_2.head(15)
```
