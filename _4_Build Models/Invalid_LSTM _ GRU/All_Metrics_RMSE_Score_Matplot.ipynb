```python
import pandas as pd
import numpy as np
#import matplotlib
#import matplotlib.pyplot as plt
#import seaborn as sns
```

```python
#讀3個檔, 設為muitiindex, 只取RMSE值
A = pd.read_csv('/content/drive/MyDrive/Final Project/LSTM/v7 data/LSTM_only5/researchdata/All_Metrics_Score_Time.csv').set_index(['name','modelname','split']).RMSE.xs('Test', level = 'split').to_frame().rename(columns = {'RMSE':'A_RMSE'})
B = pd.read_csv('/content/drive/MyDrive/Final Project/LSTM/v7 data/LSTM_TAlib15/researchdata/All_Metrics_Score_Time.csv').set_index(['name','modelname','split']).RMSE.xs('Test', level = 'split').to_frame().rename(columns = {'RMSE':'B_RMSE'})
C = pd.read_csv('/content/drive/MyDrive/Final Project/LSTM/v7 data/LSTM_Future/researchdata/All_Metrics_Score_Time.csv').set_index(['name','modelname','split']).RMSE.xs('Test', level = 'split').to_frame().rename(columns = {'RMSE':'C_RMSE'})
#3種模型RMSE值合併為一個dataframe
ALL_rmse = pd.concat([A ,B, C], axis = 1)
# 畫圖
final_chart = ALL_rmse.unstack(level=1).plot(kind='bar', title = 'ALL_rmse', figsize = (20,6), fontsize='xx-large').legend(loc = 2)
final_chart
```
